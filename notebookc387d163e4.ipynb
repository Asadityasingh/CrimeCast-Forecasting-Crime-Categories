```python
# import pandas as pd
# import numpy as np

# def preprocess_data(df):
#     """
#     Function to preprocess the crime dataset by handling missing values and creating new features.
#     Args:
#     df (pd.DataFrame): DataFrame containing the dataset.
    
#     Returns:
#     pd.DataFrame: Preprocessed DataFrame.
#     """
#     # 1. Cross_Street: Replace missing values with 'Unknown'
#     df['Cross_Street'].fillna('Unknown', inplace=True)
    
#     # 2. Modus_Operandi: Replace missing values with 'None' and create a binary feature
#     df['Modus_Operandi'].fillna('None', inplace=True)
#     df['Has_Modus_Operandi'] = np.where(df['Modus_Operandi'] == 'None', 0, 1)
    
#     # 3. Victim_Sex: Replace missing values with 'Unknown'
#     df['Victim_Sex'].fillna('Unknown', inplace=True)
    
#     # 4. Victim_Descent: Replace missing values with 'Unknown'
#     df['Victim_Descent'].fillna('Unknown', inplace=True)
    
#     # 5. Premise_Description: Replace missing values with 'Unknown'
#     df['Premise_Description'].fillna('Unknown', inplace=True)
    
#     # 6. Weapon_Used_Code and Weapon_Description: Replace missing values with 'No Weapon'
#     df['Weapon_Used_Code'].fillna('No Weapon', inplace=True)
#     df['Weapon_Description'].fillna('No Weapon', inplace=True)
    
#     return df


# # Apply preprocessing function to both train and test datasets
# train = preprocess_data(train)
# test = preprocess_data(test)

# # Save the cleaned datasets
# # train_cleaned.to_csv('cleaned_train.csv', index=False)
# # test_cleaned.to_csv('cleaned_test.csv', index=False)

# # Optional: Check for remaining missing values
# # print(train.isnull().sum()[train_cleaned.isnull().sum() > 0])
# # print(test_cleaned.isnull().sum()[test_cleaned.isnull().sum() > 0])


from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np

def preprocess_data(df):
    """
    Function to preprocess the crime dataset by handling missing values and creating new features.
    Args:
    df (pd.DataFrame): DataFrame containing the dataset.
    
    Returns:
    pd.DataFrame: Preprocessed DataFrame.
    """
    # Imputer for categorical columns
    imputer_categorical = SimpleImputer(strategy='most_frequent')

    # Imputer for numerical columns (example: using median)
    imputer_numerical = SimpleImputer(strategy='median')
    
    # 1. Cross_Street: Replace missing values with 'Unknown'
    df['Cross_Street'] = imputer_categorical.fit_transform(df[['Cross_Street']]).ravel()
    
    # 2. Modus_Operandi: Replace missing values with 'None' and create a binary feature
    df['Modus_Operandi'] = imputer_categorical.fit_transform(df[['Modus_Operandi']]).ravel()
    df['Has_Modus_Operandi'] = np.where(df['Modus_Operandi'] == 'None', 0, 1)
    
    # 3. Victim_Sex: Replace missing values with 'Unknown'
    df['Victim_Sex'] = imputer_categorical.fit_transform(df[['Victim_Sex']]).ravel()
    
    # 4. Victim_Descent: Replace missing values with 'Unknown'
    df['Victim_Descent'] = imputer_categorical.fit_transform(df[['Victim_Descent']]).ravel()
    
    # 5. Premise_Description: Replace missing values with 'Unknown'
    df['Premise_Description'] = imputer_categorical.fit_transform(df[['Premise_Description']]).ravel()
    
    # 6. Weapon_Used_Code and Weapon_Description: Replace missing values with 'No Weapon'
    df['Weapon_Used_Code'] = imputer_categorical.fit_transform(df[['Weapon_Used_Code']]).ravel()
    df['Weapon_Description'] = imputer_categorical.fit_transform(df[['Weapon_Description']]).ravel()
    
    # 7. Numerical columns
    numerical_columns = df.select_dtypes(include=[np.number]).columns
    df[numerical_columns] = imputer_numerical.fit_transform(df[numerical_columns])
    
    return df

# Apply the preprocessing function
train = preprocess_data(train)
test = preprocess_data(test)
```

```python
def extract_date_features(df):
    """
    Function to convert date columns to datetime and extract features such as year, month, and day.
    
    Args:
    df (pd.DataFrame): DataFrame containing the dataset.
    
    Returns:
    pd.DataFrame: DataFrame with extracted date features.
    """
    # Convert Date_Reported and Date_Occurred to datetime
    df['Date_Reported'] = pd.to_datetime(df['Date_Reported'], errors='coerce')
    df['Date_Occurred'] = pd.to_datetime(df['Date_Occurred'], errors='coerce')
    
    # Extracting features from Date_Occurred
    df['Year_Occurred'] = df['Date_Occurred'].dt.year
    df['Month_Occurred'] = df['Date_Occurred'].dt.month
    df['Day_Occurred'] = df['Date_Occurred'].dt.day
    df['Weekday_Occurred'] = df['Date_Occurred'].dt.weekday
    
    # Extracting features from Date_Reported
    df['Year_Reported'] = df['Date_Reported'].dt.year
    df['Month_Reported'] = df['Date_Reported'].dt.month
    df['Day_Reported'] = df['Date_Reported'].dt.day
    df['Weekday_Reported'] = df['Date_Reported'].dt.weekday
    
    df.drop(['Date_Reported', 'Date_Occurred'], axis=1, inplace=True)
    
    
    return df
```

```python
def encode_features(df):
    """
    Function to encode categorical features with specific strategies for each feature.
    
    Args:
    df (pd.DataFrame): DataFrame containing the dataset.
    
    Returns:
    pd.DataFrame: DataFrame with encoded categorical features.
    """
    # High cardinality columns handling
    high_cardinality_cols = ['Premise_Description', 'Weapon_Used_Code', 'Weapon_Description','Location', 'Cross_Street']
    other_categorical_cols = ['Area_Name', 'Victim_Sex', 'Victim_Descent','Status', 'Status_Description']
    
    # Frequency Encoding for high cardinality columns
    for col in high_cardinality_cols:
        top_n = 50  # Adjust as needed
        top_n_categories = df[col].value_counts().index[:top_n]
        df[col] = df[col].apply(lambda x: x if x in top_n_categories else 'Other')
        df[col] = pd.factorize(df[col])[0]  # Convert to integer encoding
    
    # One-Hot Encoding for other categorical columns
    df = pd.get_dummies(df, columns=other_categorical_cols, drop_first=True)
    
    return df
```

```python
def encode_modus_operandi(df):
    """
    Apply frequency encoding to the 'Modus_Operandi' column.
    
    Args:
    df (pd.DataFrame): DataFrame containing the dataset.
    
    Returns:
    pd.DataFrame: DataFrame with 'Modus_Operandi' column frequency encoded.
    """
    # Frequency Encoding for Modus_Operandi
    freq_encoding = df['Modus_Operandi'].value_counts(normalize=True)
    df['Modus_Operandi'] = df['Modus_Operandi'].map(freq_encoding)
    return df
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
import pickle
def encode_target(df, target_column):
    """
    Function to encode the target variable using Label Encoding and save the encoder.
    
    Args:
    df (pd.DataFrame): DataFrame containing the dataset.
    target_column (str): The name of the target column to encode.
    
    Returns:
    pd.DataFrame: DataFrame with encoded target variable.
    """
    label_encoder = LabelEncoder()
    df[target_column] = label_encoder.fit_transform(df[target_column])
    
    # Save the LabelEncoder object to a file
    with open('label_encoder.pkl', 'wb') as file:
        pickle.dump(label_encoder, file)
    
    return df
train = encode_target(train, 'Crime_Category')
```

```python
# Fit the models
rf_model.fit(X, y)
et_model.fit(X, y)

# Get OOB scores
rf_oob_score = rf_model.oob_score_
et_oob_score = et_model.oob_score_

# Normalize OOB scores to use as weights
total_oob = rf_oob_score + et_oob_score
rf_weight = rf_oob_score / total_oob
et_weight = et_oob_score / total_oob

# Predict probabilities
rf_proba = rf_model.predict_proba(X)
et_proba = et_model.predict_proba(X)

# Blending predictions using weighted average
blended_proba = (rf_proba * rf_weight) + (et_proba * et_weight)

# Convert probabilities to class predictions
y_pred = np.argmax(blended_proba, axis=1)

# Evaluate accuracy
accuracy = accuracy_score(y, y_pred)
print(f"Blended Model Accuracy: {accuracy}")
```
